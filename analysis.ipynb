```python
# Load the dataset
def preprocess_cloudburst_data(file_path='cloudburst.csv', test_size=0.2, random_state=42):
    # Read the data
    df = pd.read_csv(file_path)

    # Display basic information
    print(f"Dataset shape: {df.shape}")
    print("\nSample data:")
    print(df.head())

    # Check for missing values
    missing_values = df.isnull().sum()
    print("\nMissing values per column:")
    print(missing_values[missing_values > 0])

    # Handle date column - extract features from date
    # Fix: specify dayfirst=True to correctly parse DD-MM-YYYY format
    df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['DayOfWeek'] = df['Date'].dt.dayofweek
    df['Season'] = ((df['Month'] % 12) // 3 + 1).map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})

    # Encode categorical variables
    categorical_cols = ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'Season']
    label_encoders = {}

    for col in categorical_cols:
        if col in df.columns:
            # Replace missing values with 'Unknown'
            df[col].fillna('Unknown', inplace=True)

            # Apply label encoding
            le = LabelEncoder()
            df[col + '_encoded'] = le.fit_transform(df[col])
            label_encoders[col] = le

    # Handle missing numerical values
    numerical_cols = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
                      'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm',
                      'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm',
                      'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']

    for col in numerical_cols:
        if col in df.columns:
            # Fill missing values with median
            df[col].fillna(df[col].median(), inplace=True)

    # Prepare target variables
    # Assuming 'CloudBurst Today' and 'CloudBurstTomorrow' are the target variables
    # and they contain 'Yes'/'No' or 1/0 values

    target_cols = ['CloudBurst Today', 'CloudBurstTomorrow']
    for col in target_cols:
        if col in df.columns:
            if df[col].dtype == 'object':
                # Convert 'Yes'/'No' to 1/0
                df[col] = df[col].map({'Yes': 1, 'No': 0})
            # Ensure any remaining NaN are filled with the most common value
            df[col].fillna(df[col].mode()[0], inplace=True)

    # Select features for the model
    feature_cols = [
        'MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
        'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm',
        'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm',
        'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm',
        'Year', 'Month', 'Day', 'DayOfWeek',
        'Location_encoded', 'WindGustDir_encoded',
        'WindDir9am_encoded', 'WindDir3pm_encoded', 'Season_encoded'
    ]

    # Ensure all feature columns exist in the dataframe
    features = [col for col in feature_cols if col in df.columns]
    X = df[features]

    # Choose target (CloudBurst Today or CloudBurstTomorrow)
    target = 'CloudBurstTomorrow'  # Change this if needed
    if target in df.columns:
        y = df[target]
    else:
        print(f"Target variable '{target}' not found in dataframe.")
        return None, None, None

    # Normalize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=test_size, random_state=random_state
    )

    print(f"\nTraining set shape: {X_train.shape}")
    print(f"Testing set shape: {X_test.shape}")

    # Create a dictionary to store preprocessing objects
    preprocessing_info = {
        'feature_columns': features,
        'label_encoders': label_encoders,
        'scaler': scaler,
        'target_column': target
    }

    return X_train, X_test, y_train, y_test, preprocessing_info

# Example usage
# X_train, X_test, y_train, y_test, preprocessing_info = preprocess_cloudburst_data('cloudburst.csv')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
# Function to fit a linear regression model
def fit_linear_regression(X_train, X_test, y_train, y_test, preprocessing_info):
    # Initialize the Linear Regression model
    model = LinearRegression()

    # Fit the model
    print("Fitting Linear Regression model...")
    model.fit(X_train, y_train)

    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Round predictions for binary classification (since the target is 0/1)
    y_train_pred_rounded = np.round(y_train_pred)
    y_test_pred_rounded = np.round(y_test_pred)

    # Calculate metrics
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)

    train_mae = mean_absolute_error(y_train, y_train_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)

    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    # Calculate accuracy for binary classification
    train_accuracy = np.mean(y_train_pred_rounded == y_train)
    test_accuracy = np.mean(y_test_pred_rounded == y_test)

    # Print metrics
    print("\nLinear Regression Model Performance:")
    print(f"Training MSE: {train_mse:.4f}")
    print(f"Testing MSE: {test_mse:.4f}")
    print(f"Training MAE: {train_mae:.4f}")
    print(f"Testing MAE: {test_mae:.4f}")
    print(f"Training R: {train_r2:.4f}")
    print(f"Testing R: {test_r2:.4f}")
    print(f"Training Accuracy: {train_accuracy:.4f}")
    print(f"Testing Accuracy: {test_accuracy:.4f}")

    # Analyze coefficients
    feature_names = preprocessing_info['feature_columns']
    coefficients = pd.DataFrame(model.coef_, index=feature_names, columns=['Coefficient'])

    # Sort coefficients by absolute value
    coefficients['Abs_Coefficient'] = coefficients['Coefficient'].abs()
    sorted_coefficients = coefficients.sort_values(by='Abs_Coefficient', ascending=False)

    print("\nTop 10 most important features:")
    print(sorted_coefficients.head(10))

    # Plot actual vs predicted values
    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, y_test_pred, alpha=0.5)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title('Actual vs Predicted Values (Test Set)')
    plt.tight_layout()
    plt.show()

    # Plot feature importance
    plt.figure(figsize=(12, 8))
    sorted_coefficients_plot = sorted_coefficients.head(15)  # Top 15 features
    sns.barplot(x='Coefficient', y=sorted_coefficients_plot.index, data=sorted_coefficients_plot)
    plt.title('Top 15 Feature Coefficients')
    plt.xlabel('Coefficient Value')
    plt.tight_layout()
    plt.show()

    return model, sorted_coefficients

# Main execution
if __name__ == "__main__":
    # Preprocess the data
    X_train, X_test, y_train, y_test, preprocessing_info = preprocess_cloudburst_data('cloudburst.csv')

    # Fit linear regression model
    model, coefficients = fit_linear_regression(X_train, X_test, y_train, y_test, preprocessing_info)

    print("\nCompleted linear regression modeling.")
```

```python
def create_test_sample(date, location, mintemp, maxtemp, rainfall, evaporation, sunshine, windgustdir, windgustspeed, winddir3pm, winddir9am, windspeed9am,
                       windspeed3pm, humidity9am, humidity3pm, pressure9am, pressure3pm, cloud9am, cloud3pm, temp9am, temp3pm, description):
    """
    Creates a single synthetic test data sample for the cloudburst prediction model.

    Returns:
    pandas DataFrame with one test sample
    """
    # Create a DataFrame with one sample weather scenario
    test_sample = pd.DataFrame([
        # Rainy day with low pressure (likely cloudburst)
        {
            'Date': date,
            'Location': location,
            'MinTemp': mintemp,
            'MaxTemp': maxtemp,
            'Rainfall': rainfall,
            'Evaporation': evaporation,
            'Sunshine': sunshine,
            'WindGustDir': windgustdir,
            'WindGustSpeed': windgustspeed,
            'WindDir9am': winddir3pm,
            'WindDir3pm': winddir9am,
            'WindSpeed9am': windspeed9am,
            'WindSpeed3pm': windspeed3pm,
            'Humidity9am': humidity9am,
            'Humidity3pm': humidity3pm,
            'Pressure9am': pressure9am,
            'Pressure3pm': pressure3pm,
            'Cloud9am': cloud9am,
            'Cloud3pm': cloud3pm,
            'Temp9am': temp9am,
            'Temp3pm': temp3pm,
            'Description': description
        }
    ])

    # Convert Date to datetime
    test_sample['Date'] = pd.to_datetime(test_sample['Date'], dayfirst=True)

    return test_sample

def predict_on_sample(model, preprocessing_info, test_sample):
    """
    Make prediction using the trained model on a single test sample

    Parameters:
    model : trained model (Linear Regression, Random Forest, etc.)
    preprocessing_info : dictionary with preprocessing details
    test_sample : DataFrame with a single test sample

    Returns:
    DataFrame with prediction added
    """
    # Add date-derived features
    test_sample['Year'] = test_sample['Date'].dt.year
    test_sample['Month'] = test_sample['Date'].dt.month
    test_sample['Day'] = test_sample['Date'].dt.day
    test_sample['DayOfWeek'] = test_sample['Date'].dt.dayofweek
    test_sample['Season'] = ((test_sample['Month'] % 12) // 3 + 1).map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})

    # Encode categorical variables using the provided label encoders
    label_encoders = preprocessing_info['label_encoders']

    for col, encoder in label_encoders.items():
        # Fill missing values
        test_sample[col].fillna('Unknown', inplace=True)

        # Transform values that exist in the training data
        test_sample[col + '_encoded'] = test_sample[col].apply(
            lambda x: encoder.transform([x])[0] if x in encoder.classes_ else -1
        )

    # Get feature columns and scaler
    features = preprocessing_info['feature_columns']
    scaler = preprocessing_info['scaler']

    # Ensure all feature columns exist in the test sample
    for feature in features:
        if feature not in test_sample.columns:
            print(f"Warning: Feature '{feature}' not found in test sample. Setting to 0.")
            test_sample[feature] = 0

    # Prepare feature matrix
    X_test = test_sample[features]

    # Scale features
    X_test_scaled = scaler.transform(X_test)

    # Make predictions
    prediction = model.predict(X_test_scaled)[0]
    test_sample['Predicted_Raw'] = prediction
    test_sample['Predicted_Rounded'] = round(prediction)

    # Print results
    print("\nPrediction for test sample:")
    print(f"Date: {test_sample['Date'].iloc[0].strftime('%d-%m-%Y')}")
    print(f"Location: {test_sample['Location'].iloc[0]}")
    print(f"Weather: {test_sample['Description'].iloc[0]}")
    print(f"Key weather indicators:")
    print(f"  - Rainfall: {test_sample['Rainfall'].iloc[0]} mm")
    print(f"  - Humidity: {test_sample['Humidity3pm'].iloc[0]}%")
    print(f"  - Wind speed: {test_sample['WindGustSpeed'].iloc[0]} km/h")
    print(f"  - Pressure: {test_sample['Pressure3pm'].iloc[0]} hPa")
    print(f"  - Cloud cover: {test_sample['Cloud3pm'].iloc[0]}/8")
    print(f"\nCloudburst prediction: {prediction:.4f}")
    print(f"Classification: {'CloudBurst' if round(prediction) == 1 else 'No CloudBurst'}")

    return test_sample
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Preprocess data
def preprocess_data(file_path='cloudburst.csv', test_size=0.2, random_state=42):
    # Read the data
    df = pd.read_csv(file_path)

    # Display basic information
    print(f"Dataset shape: {df.shape}")

    # Check for missing values
    missing_values = df.isnull().sum()
    print("\nMissing values per column:")
    print(missing_values[missing_values > 0])

    # Handle date column - extract features from date
    df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['DayOfWeek'] = df['Date'].dt.dayofweek
    df['Season'] = ((df['Month'] % 12) // 3 + 1).map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})

    # Encode categorical variables
    categorical_cols = ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'Season']
    label_encoders = {}

    for col in categorical_cols:
        if col in df.columns:
            # Replace missing values with 'Unknown'
            df[col].fillna('Unknown', inplace=True)

            # Apply label encoding
            le = LabelEncoder()
            df[col + '_encoded'] = le.fit_transform(df[col])
            label_encoders[col] = le

    # Handle missing numerical values
    numerical_cols = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
                      'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm',
                      'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm',
                      'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']

    for col in numerical_cols:
        if col in df.columns:
            # Fill missing values with median
            df[col].fillna(df[col].median(), inplace=True)

    # Prepare target variables
    target_cols = ['CloudBurst Today', 'CloudBurstTomorrow']
    for col in target_cols:
        if col in df.columns:
            if df[col].dtype == 'object':
                # Convert 'Yes'/'No' to 1/0
                df[col] = df[col].map({'Yes': 1, 'No': 0})
            # Ensure any remaining NaN are filled with the most common value
            df[col].fillna(df[col].mode()[0], inplace=True)

    # Select features for the model
    feature_cols = [
        'MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
        'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm',
        'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm',
        'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm',
        'Year', 'Month', 'Day', 'DayOfWeek',
        'Location_encoded', 'WindGustDir_encoded',
        'WindDir9am_encoded', 'WindDir3pm_encoded', 'Season_encoded'
    ]

    # Ensure all feature columns exist in the dataframe
    features = [col for col in feature_cols if col in df.columns]
    X = df[features]

    # Choose target (CloudBurst Today or CloudBurstTomorrow)
    target = 'CloudBurstTomorrow'  # Change this if needed
    if target in df.columns:
        y = df[target]
    else:
        print(f"Target variable '{target}' not found in dataframe.")
        return None, None, None, None, None

    # Normalize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=test_size, random_state=random_state, stratify=y
    )

    print(f"\nTraining set shape: {X_train.shape}")
    print(f"Testing set shape: {X_test.shape}")

    # Create a dictionary to store preprocessing objects
    preprocessing_info = {
        'feature_columns': features,
        'label_encoders': label_encoders,
        'scaler': scaler,
        'target_column': target,
        'input_size': len(features)
    }

    return X_train, X_test, y_train, y_test, preprocessing_info

# Create a neural network model
def create_model(input_size, hidden_size=64, dropout_rate=0.3):
    model = nn.Sequential(
        nn.Linear(input_size, hidden_size),
        nn.BatchNorm1d(hidden_size),
        nn.ReLU(),
        nn.Dropout(dropout_rate),

        nn.Linear(hidden_size, hidden_size),
        nn.BatchNorm1d(hidden_size),
        nn.ReLU(),
        nn.Dropout(dropout_rate),

        nn.Linear(hidden_size, 32),
        nn.BatchNorm1d(32),
        nn.ReLU(),
        nn.Dropout(dropout_rate),

        nn.Linear(32, 1),
        nn.Sigmoid()
    )
    return model

# Train neural network
def train_neural_network(X_train, y_train, preprocessing_info, batch_size=64, learning_rate=0.001,
                       hidden_size=64, epochs=100, dropout_rate=0.3, val_split=0.1):
    # Set device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    # Create validation split
    X_train_arr, X_val_arr, y_train_arr, y_val_arr = train_test_split(
        X_train, y_train, test_size=val_split, random_state=42, stratify=y_train
    )

    # Convert to PyTorch tensors
    X_train_tensor = torch.tensor(X_train_arr, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train_arr.values, dtype=torch.float32).reshape(-1, 1)
    X_val_tensor = torch.tensor(X_val_arr, dtype=torch.float32)
    y_val_tensor = torch.tensor(y_val_arr.values, dtype=torch.float32).reshape(-1, 1)

    # Create datasets and data loaders
    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    val_dataset = TensorDataset(X_val_tensor, y_val_tensor)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)

    # Initialize model
    input_size = preprocessing_info['input_size']
    model = create_model(input_size, hidden_size, dropout_rate).to(device)

    # Define loss function and optimizer
    criterion = nn.BCELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-5)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=True)

    # Training loop
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    best_model_state = None

    for epoch in range(epochs):
        # Training phase
        model.train()
        train_loss = 0.0
        for inputs, targets in train_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            # Zero gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(inputs)

            # Calculate loss
            loss = criterion(outputs, targets)

            # Backward pass and update
            loss.backward()
            optimizer.step()

            train_loss += loss.item() * inputs.size(0)

        train_loss /= len(train_loader.dataset)
        train_losses.append(train_loss)

        # Validation phase
        model.eval()
        val_loss = 0.0
        val_preds = []
        val_targets = []

        with torch.no_grad():
            for inputs, targets in val_loader:
                inputs, targets = inputs.to(device), targets.to(device)

                # Forward pass
                outputs = model(inputs)

                # Calculate loss
                loss = criterion(outputs, targets)
                val_loss += loss.item() * inputs.size(0)

                # Store predictions and targets
                val_preds.extend(outputs.cpu().numpy())
                val_targets.extend(targets.cpu().numpy())

        val_loss /= len(val_loader.dataset)
        val_losses.append(val_loss)

        # Adjust learning rate
        scheduler.step(val_loss)

        # Save best model
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_state = model.state_dict().copy()

        # Print progress
        if (epoch + 1) % 10 == 0:
            # Calculate accuracy
            val_preds_binary = np.round(val_preds)
            val_accuracy = np.mean(val_preds_binary == np.array(val_targets))

            print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.4f}, "
                  f"Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}")

    # Load best model
    model.load_state_dict(best_model_state)

    # Plot training curves
    plt.figure(figsize=(10, 6))
    plt.plot(train_losses, label='Training Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

    return model, device

# Evaluate function
def evaluate_model(model, X_test, y_test, device):
    # Convert to PyTorch tensors
    X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
    y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).reshape(-1, 1)

    # Create test dataset and data loader
    test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
    test_loader = DataLoader(test_dataset, batch_size=128)

    # Evaluation
    model.eval()
    y_pred = []
    y_true = []

    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)

            y_pred.extend(outputs.cpu().numpy())
            y_true.extend(targets.cpu().numpy())

    # Convert to numpy arrays
    y_pred = np.array(y_pred).flatten()
    y_true = np.array(y_true).flatten()

    # Compute metrics
    y_pred_binary = np.round(y_pred)
    accuracy = np.mean(y_pred_binary == y_true)

    from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve

    cm = confusion_matrix(y_true, y_pred_binary)
    print("\nConfusion Matrix:")
    print(cm)

    print("\nClassification Report:")
    print(classification_report(y_true, y_pred_binary))

    # Calculate ROC AUC
    auc = roc_auc_score(y_true, y_pred)
    print(f"ROC AUC: {auc:.4f}")

    # Plot ROC curve
    fpr, tpr, _ = roc_curve(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'AUC = {auc:.4f}')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.grid(True)
    plt.show()

    return y_pred, y_true, accuracy

# Function to predict on a single sample
def predict_sample(model, preprocessing_info, device):
    # Create sample data
    sample_data = {
        'Date': '05-06-2023',
        'Location': 'Brisbane',
        'MinTemp': 20.3,
        'MaxTemp': 23.5,
        'Rainfall': 45.2,
        'Evaporation': 1.8,
        'Sunshine': 0.0,
        'WindGustDir': 'SSE',
        'WindGustSpeed': 83.0,
        'WindDir9am': 'S',
        'WindDir3pm': 'SSE',
        'WindSpeed9am': 19.0,
        'WindSpeed3pm': 24.0,
        'Humidity9am': 95.0,
        'Humidity3pm': 92.0,
        'Pressure9am': 998.2,
        'Pressure3pm': 997.5,
        'Cloud9am': 8.0,
        'Cloud3pm': 8.0,
        'Temp9am': 21.5,
        'Temp3pm': 22.8
    }

    # Convert to DataFrame
    sample_df = pd.DataFrame([sample_data])
    sample_df['Date'] = pd.to_datetime(sample_df['Date'], dayfirst=True)

    # Add derived features
    sample_df['Year'] = sample_df['Date'].dt.year
    sample_df['Month'] = sample_df['Date'].dt.month
    sample_df['Day'] = sample_df['Date'].dt.day
    sample_df['DayOfWeek'] = sample_df['Date'].dt.dayofweek
    sample_df['Season'] = ((sample_df['Month'] % 12) // 3 + 1).map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})

    # Encode categorical variables
    for col, encoder in preprocessing_info['label_encoders'].items():
        sample_df[col].fillna('Unknown', inplace=True)
        sample_df[col + '_encoded'] = sample_df[col].apply(
            lambda x: encoder.transform([x])[0] if x in encoder.classes_ else -1
        )

    # Prepare features
    features = preprocessing_info['feature_columns']
    for feature in features:
        if feature not in sample_df.columns:
            sample_df[feature] = 0

    X_sample = sample_df[features].values

    # Scale features
    X_sample_scaled = preprocessing_info['scaler'].transform(X_sample)

    # Convert to tensor
    X_sample_tensor = torch.tensor(X_sample_scaled, dtype=torch.float32).to(device)

    # Make prediction
    model.eval()
    with torch.no_grad():
        prediction = model(X_sample_tensor).cpu().numpy()[0][0]

    # Print results
    print("\nNeural Network Prediction on Sample Data:")
    print(f"Weather: Rainy day with low pressure in Brisbane")
    print(f"Key indicators: Rainfall={sample_data['Rainfall']}mm, Humidity={sample_data['Humidity3pm']}%, Pressure={sample_data['Pressure3pm']}hPa")
    print(f"\nCloudburst Probability: {prediction:.4f}")
    print(f"Classification: {'CloudBurst' if prediction >= 0.5 else 'No CloudBurst'}")

    return prediction

# Main function to run everything
def main():
    # Preprocess data
    X_train, X_test, y_train, y_test, preprocessing_info = preprocess_data('cloudburst.csv')

    # Train model
    model, device = train_neural_network(
        X_train, y_train, preprocessing_info,
        batch_size=64,
        learning_rate=0.0001,
        hidden_size=64,
        epochs=100,
        dropout_rate=0.3
    )

    # Evaluate model
    y_pred, y_true, accuracy = evaluate_model(model, X_test, y_test, device)
    print(f"\nTest Accuracy: {accuracy:.4f}")

    # Predict on sample
    sample_prediction = predict_sample(model, preprocessing_info, device)

    # Save model
    torch.save({
        'model_state_dict': model.state_dict(),
        'preprocessing_info': preprocessing_info
    }, 'cloudburst_nn_model.pth')

    print("\nModel saved as 'cloudburst_nn_model.pth'")

# Call the main function if script is run directly
if __name__ == "__main__":
    main()
```
